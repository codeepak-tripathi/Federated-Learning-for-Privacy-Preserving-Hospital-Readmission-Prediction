```python
import pandas as pd
import numpy as np
import kagglehub

# Ensure 'path' points to the downloaded dataset directory.
# Re-calling dataset_download to guarantee 'path' is correctly set in this cell's scope.
path = kagglehub.dataset_download("faizan729/syntheticmass-synthea-synthetic-ehr-dataset")

patients     = pd.read_csv(f"{path}/patients.csv")
conditions   = pd.read_csv(f"{path}/conditions.csv")
encounters   = pd.read_csv(f"{path}/encounters.csv")
observations = pd.read_csv(f"{path}/observations.csv")

# Convert dates
encounters['START'] = pd.to_datetime(encounters['START'])
encounters['STOP']  = pd.to_datetime(encounters['STOP'])

# Sort by patient and date
encounters = encounters.sort_values(['PATIENT', 'START'])

# Calculate days between consecutive visits per patient
encounters['NEXT_VISIT'] = encounters.groupby('PATIENT')['START'].shift(-1)
encounters['DAYS_TO_NEXT'] = (encounters['NEXT_VISIT'] - encounters['STOP']).dt.days

# Target: readmitted within 30 days
encounters['READMITTED_30'] = (encounters['DAYS_TO_NEXT'] <= 30).astype(int)

print("Readmission distribution:")
print(encounters['READMITTED_30'].value_counts())
print("\nReadmission rate:", encounters['READMITTED_30'].mean().round(3))

print("\nEncounters sample with target:")
print(encounters[['PATIENT','START','STOP','DAYS_TO_NEXT','READMITTED_30']].head(10))
```

```python
# --- PATIENT FEATURES ---
patient_features = patients[['Id', 'BIRTHDATE', 'GENDER', 'RACE', 'ETHNICITY', 'DEATHDATE']].copy()
patient_features['BIRTHDATE'] = pd.to_datetime(patient_features['BIRTHDATE'])

# Age at time of encounter
encounters = encounters.merge(patient_features, left_on='PATIENT', right_on='Id', how='left')
encounters['AGE'] = (encounters['START'] - encounters['BIRTHDATE'].dt.tz_localize('UTC')).dt.days // 365
encounters['IS_DEAD'] = encounters['DEATHDATE'].notna().astype(int)

# --- ENCOUNTER FEATURES ---
encounters['ENCOUNTER_DURATION_MINS'] = (encounters['STOP'] - encounters['START']).dt.total_seconds() / 60
encounters['ENCOUNTER_TYPE'] = encounters['ENCOUNTERCLASS']

# --- CONDITION COUNT PER PATIENT ---
condition_counts = conditions.groupby('PATIENT')['DESCRIPTION'].count().reset_index()
condition_counts.columns = ['PATIENT', 'CONDITION_COUNT']
encounters = encounters.merge(condition_counts, on='PATIENT', how='left')

# --- FINAL FEATURE SET ---
feature_cols = [
    'AGE', 'GENDER', 'RACE', 'ETHNICITY',
    'ENCOUNTER_DURATION_MINS', 'ENCOUNTER_TYPE',
    'CONDITION_COUNT', 'IS_DEAD', 'BASE_ENCOUNTER_COST',
    'TOTAL_CLAIM_COST', 'PAYER_COVERAGE'
]

df = encounters[feature_cols + ['PATIENT', 'READMITTED_30']].dropna()

print("Final dataset shape:", df.shape)
print("\nFeature sample:")
print(df[feature_cols].head(5))
print("\nClass distribution:")
print(df['READMITTED_30'].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# --- ENCODE CATEGORICAL COLUMNS ---
cat_cols = ['GENDER', 'RACE', 'ETHNICITY', 'ENCOUNTER_TYPE']
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col].astype(str))

# --- FINAL FEATURES & TARGET ---
X = df[feature_cols].values
y = df['READMITTED_30'].values
patients_ids = df['PATIENT'].values

# --- SPLIT INTO 3 HOSPITALS BY PATIENT ---
unique_patients = df['PATIENT'].unique()
np.random.seed(42)
np.random.shuffle(unique_patients)

h1_patients = unique_patients[:len(unique_patients)//3]
h2_patients = unique_patients[len(unique_patients)//3:2*len(unique_patients)//3]
h3_patients = unique_patients[2*len(unique_patients)//3:]

hospital_1 = df[df['PATIENT'].isin(h1_patients)]
hospital_2 = df[df['PATIENT'].isin(h2_patients)]
hospital_3 = df[df['PATIENT'].isin(h3_patients)]

print("Hospital 1 size:", hospital_1.shape)
print("Hospital 2 size:", hospital_2.shape)
print("Hospital 3 size:", hospital_3.shape)

print("\nHospital 1 readmission rate:", hospital_1['READMITTED_30'].mean().round(3))
print("Hospital 2 readmission rate:", hospital_2['READMITTED_30'].mean().round(3))
print("Hospital 3 readmission rate:", hospital_3['READMITTED_30'].mean().round(3))
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.preprocessing import StandardScaler
from copy import deepcopy

# --- PREP EACH HOSPITAL DATA ---
def get_xy(hospital_df):
    X = hospital_df[feature_cols].values
    y = hospital_df['READMITTED_30'].values
    return X, y

X1, y1 = get_xy(hospital_1)
X2, y2 = get_xy(hospital_2)
X3, y3 = get_xy(hospital_3)

# Scale each hospital independently (they dont share data)
scaler1, scaler2, scaler3 = StandardScaler(), StandardScaler(), StandardScaler()
X1 = scaler1.fit_transform(X1)
X2 = scaler2.fit_transform(X2)
X3 = scaler3.fit_transform(X3)

# --- LOCAL TRAINING FUNCTION ---
def train_local(X, y):
    model = LogisticRegression(max_iter=1000)
    model.fit(X, y)
    return model

# --- FEDAVG AGGREGATION ---
def fedavg(models, weights):
    """Average model coefficients weighted by hospital size"""
    total = sum(weights)
    avg_coef = sum(m.coef_ * (w/total) for m, w in zip(models, weights))
    avg_intercept = sum(m.intercept_ * (w/total) for m, w in zip(models, weights))

    global_model = deepcopy(models[0])
    global_model.coef_ = avg_coef
    global_model.intercept_ = avg_intercept
    return global_model

# --- FEDERATED TRAINING LOOP ---
print("=== Federated Learning Training ===\n")

NUM_ROUNDS = 5

for round_num in range(1, NUM_ROUNDS + 1):
    # Each hospital trains locally
    m1 = train_local(X1, y1)
    m2 = train_local(X2, y2)
    m3 = train_local(X3, y3)

    # Aggregate via FedAvg
    global_model = fedavg(
        models=[m1, m2, m3],
        weights=[len(X1), len(X2), len(X3)]
    )

    # Evaluate global model on each hospital
    auc1 = roc_auc_score(y1, global_model.predict_proba(X1)[:,1])
    auc2 = roc_auc_score(y2, global_model.predict_proba(X2)[:,1])
    auc3 = roc_auc_score(y3, global_model.predict_proba(X3)[:,1])
    avg_auc = np.mean([auc1, auc2, auc3])

    print(f"Round {round_num} | H1 AUC: {auc1:.3f} | H2 AUC: {auc2:.3f} | H3 AUC: {auc3:.3f} | Avg AUC: {avg_auc:.3f}")

print("\n=== Training Complete ===")
```

```python
from sklearn.neural_network import MLPClassifier
import pickle
# --- FEDERATED TRAINING WITH MLP ---
print("=== Federated Learning (MLP) ===\n")

NUM_ROUNDS = 10
round_aucs = []

# Initialize one global model
global_coef = None
global_intercept = None

for round_num in range(1, NUM_ROUNDS + 1):
    local_models = []

    for X, y in [(X1, y1), (X2, y2), (X3, y3)]:
        model = MLPClassifier(
            hidden_layer_sizes=(64, 32),
            max_iter=10,          # only 10 iterations per round
            warm_start=False,
            random_state=42
        )
        model.fit(X, y)
        local_models.append(model)

    # Evaluate each local model on all hospitals
    auc1 = roc_auc_score(y1, local_models[0].predict_proba(X1)[:,1])
    auc2 = roc_auc_score(y2, local_models[1].predict_proba(X2)[:,1])
    auc3 = roc_auc_score(y3, local_models[2].predict_proba(X3)[:,1])
    avg_auc = np.mean([auc1, auc2, auc3])
    round_aucs.append(avg_auc)


    print(f"Round {round_num:2d} | H1: {auc1:.3f} | H2: {auc2:.3f} | H3: {auc3:.3f} | Avg: {avg_auc:.3f}")


# --- CENTRALIZED BASELINE (all data pooled) ---
print("\n=== Centralized Baseline ===")
X_all = np.vstack([X1, X2, X3])
y_all = np.concatenate([y1, y2, y3])

central_model = MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=100, random_state=42)
central_model.fit(X_all, y_all)
central_auc = roc_auc_score(y_all, central_model.predict_proba(X_all)[:,1])
print(f"Centralized AUC: {central_auc:.3f}")

with open('/content/central_model.pkl', 'wb') as f:
    pickle.dump(central_model, f)
    print(" models saved ✅")
# --- COMPARISON ---
print("\n=== FL vs Centralized ===")
print(f"FL Best AUC:          {max(round_aucs):.3f}")
print(f"Centralized AUC:      {central_auc:.3f}")
print(f"Difference:           {abs(max(round_aucs) - central_auc):.3f}")
```
